```python
life_df = pd.read_csv("life-expectancy-vs-health-expenditure.csv")
life_df.head()
```

```python
life_df = life_df[(life_df.Year<=1970) & (life_df.Year>=1920)]
life_df.head()
```

```python
life_df.columns
```

```python
life_df = life_df.groupby(['Entity', 'Code'])[['Life expectancy at birth, total (years)', 'Total population (Gapminder, HYDE & UN)']].mean().reset_index()
life_df.head()
```

```python
plt.figure(figsize = (20, 40))
sns.barplot(x='Life expectancy at birth, total (years)', y='Entity', data = life_df.sort_values('Life expectancy at birth, total (years)'))
plt.show()
```

```python
# list of countries that are observed in COVID_DF but absent in LIFE_DF
new_list = [
    list(set(covid_df.location).difference(life_df.Entity))
]
new_list
```

```python
# list of countries that are observed in life_Df but are not in COVID_DF
new_list2 = [
    list(set(life_df.Entity).difference(covid_df.location))
]

new_list2
```

```python
# Since the COVID_DF is the main dataset, we merge life_df to it. 
covid_df = pd.merge(covid_df, life_df, left_on='location', right_on='Entity')
covid_df.head()
```

```python
covid_df.shape
```

```python
covid_df.dtypes
```

```python
covid_df['log_life'] = np.log(covid_df['Life expectancy at birth, total (years)'] + 1)
covid_df['log_death'] = np.log(covid_df['total_deaths'] + 1)
```

```python
#covid_df['life_curr'] = np.log(covid_df['life_expectancy'] + 1)
```

```python
plt.figure(figsize=(10, 10))
sns.regplot(x="log_life", y='log_death', data=covid_df)
```

```python
covid_reg = covid_df[['log_death' ,'log_life','population_density', 'median_age', 'gdp_per_capita', 'aged_65_older', 'life_expectancy']]
covid_reg.dropna(inplace=True)
covid_reg.shape
```

```python
X = covid_reg[['log_life']]
y = covid_reg['log_death']
```

```python
X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```
